```python
#IPython is what you are using now to run the notebook
import IPython

# Pandas makes working with data tables easier
import pandas as pd

import json
```

```python
#load dataset
pokemon = pd.read_csv("Pokemon.csv", low_memory=False, encoding = "ISO-8859-1")
```

```python
pokemon.head()
```

Dict of Node and Link

Node: each pokemon in sorted name order - ID, name, Type

Link: source - index, target - index 

```python
ids = pokemon.ID.values.tolist()
names = pokemon.Name.values.tolist()
types = pokemon["Type 1"].values.tolist()
evolution = pokemon.Evolution.values.tolist()

dataLen = len(ids)
```

```python
nodes = []
links = []

for i in range(dataLen):
    node = {
        "id": ids[i],
        "name": names[i],
        "type": types[i]
    }
    nodes.append(node)
    
    if(not pd.isnull(evolution[i])):
        targets = evolution[i].split(",")
        
        for j in range(len(targets)):
            link = {
                "source": i,
                "target": ids.index(targets[j])
            }
            links.append(link)
```

```python
output = {
    "node":nodes,
    "link":links
}
```

```python
with open("pokemon.json","w") as dest:
    json.dump(output,dest)
```
